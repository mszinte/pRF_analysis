## Explanations
This code below is giving you an example of how to manipulate the data to get <br>
20 vertices with the best explained pRF model in V1 and then the corresponding  <br>
signals in both the MotConf and the pRF task. <br>

```python
# Define inputs
main_dir = '/home/mszinte/disks/meso_S/data/' # to change depending on how you have mount your disks 
# in the mesocentre the data are on /scratch/mszinte/data/MotConf/
project_dir = 'MotConf'
subject = 'sub-01'
group = '327'
format_ = 'fsnative'
```

```python
# Stop warnings
import warnings
warnings.filterwarnings("ignore")

# Debug
import ipdb
deb = ipdb.set_trace

# General imports
import os
import sys
import json
import numpy as np
import glob
import pandas as pd

# Personal imports
sys.path.append("{}/../utils".format(os.getcwd()))
from surface_utils import make_surface_image , load_surface
```

```python
# Get retinotopy derivatives TSV file
tsv_dir = '{}/{}/derivatives/pp_data/{}/{}/prf/tsv'.format(
    main_dir, project_dir, subject, format_)
os.makedirs(tsv_dir, exist_ok=True)

tsv_fn = '{}/{}_css-all_derivatives.tsv'.format(tsv_dir, subject)
data_derivatives = pd.read_table(tsv_fn, sep="\t")
```

```python
# get time series of selected vertices
pp_dir = "{}/{}/derivatives/pp_data".format(main_dir, project_dir)
data_dict = {}
df_roi = pd.DataFrame()
sessions = ['ses-03']
hemis = ['hemi-L', 'hemi-R']
runs = ['run-01', 'run-02'] # put all funsyou want
rois = ['V1'] # put all areas you want
num_best = 20 # number of best vertices per roi

for hemi in ['hemi-L', 'hemi-R']:
    print(hemi)
    for session in sessions:
        print(session)
        for run in runs:
            print(run)

            # define filenames
            dct_avg_gii_prf_fn = "{}/{}/fsnative/func/fmriprep_dct/{}_{}_task-pRF_{}_{}_space-fsnative_dct_bold.func.gii".format(
                pp_dir, subject, subject, session, run, hemi)
            dct_avg_gii_motconf_fn = "{}/{}/fsnative/func/fmriprep_dct/{}_{}_task-MotConf_{}_{}_space-fsnative_dct_bold.func.gii".format(
                pp_dir, subject, subject, session, run, hemi)
            
            # load surface timeseries data
            print('loading pRF data...')
            img_prf, raw_data_prf = load_surface(fn=dct_avg_gii_prf_fn)
            
            print('loading motconf data...')
            img_motconf, raw_data_motconf = load_surface(fn=dct_avg_gii_motconf_fn)

            for roi in rois:
                # get hemi half
                data_roi_hemi = data_derivatives.loc[(data_derivatives.roi == roi) & (data_derivatives.hemi == hemi)]
                data_roi_hemi.sort_values(by=['prf_loo_r2'], ascending=[False], inplace=True)
                vert_num_list = np.array(data_roi_hemi.num_vert[0:int(num_best/2)])

                print(roi)
                for vert_num in vert_num_list:
                    data_dict['roi'] = roi
                    data_dict['session'] = session
                    data_dict['run'] = run
                    data_dict['hemi'] = hemi
                    data_dict['vert_num'] = vert_num
                    data_dict['prf_bold_zscore'] = [raw_data_prf[:,vert_num]]
                    data_dict['motconf_bold_zscore'] = [raw_data_motconf[:,vert_num]]
                    df_roi = pd.concat([df_roi, pd.DataFrame(data_dict)], ignore_index=True)
```

```python
import matplotlib.pyplot as plt

# plot them simply with matplotlib
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(20, 10))
for vert_roi in np.arange(0, df_roi.loc[(df_roi.roi == 'V1')].motconf_bold_zscore.size, 1):

    ax[0].plot(df_roi.loc[(df_roi.roi == 'V1')].motconf_bold_zscore[vert_roi]);
    ax[0].set_xlim(0, 400)
    ax[0].set_xlabel('Time (s)')
    ax[0].set_ylim(-4, 4)
    ax[0].set_ylabel('BOLD z-score')
    ax[0].set_title('MotConf')

    ax[1].plot(df_roi.loc[(df_roi.roi == 'V1')].prf_bold_zscore[vert_roi]);
    ax[1].set_xlim(0, 210)
    ax[1].set_xlabel('Time (s)')
    ax[1].set_ylim(-4, 4)
    ax[1].set_ylabel('BOLD z-score')
    ax[1].set_title('pRF')
```
